```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

**Split data into features and target variable**

```python
x = breast_cancer_df.drop(columns="label", axis=1)
y = breast_cancer_df["label"]
```

**Split into data into training and testing**

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

**Building the Neural Network**

```python
# Set up layers of Neural Network
tf.random.set_seed(3)
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation = "relu"),
    keras.layers.Dense(2, activation="sigmoid")
])
```

```python
# Compiling the Neural Network
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# Training the neural network
history = model.fit(x_train, y_train, validation_split=0.1, epochs = 10)
```

**Visualize the accuracy and loss**

```python
# Accuracy plot
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.legend(["training dat", "val data"], loc="lower right") # Changed loc argument to a valid string value
plt.xlabel("Training Accuracy")
plt.ylabel("validation Accuracy")
plt.show()
```

```python
# Loss Plot
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.legend(["training dat", "val data"], loc="lower right") # Changed loc argument to a valid string value
plt.xlabel("Train Loss")
plt.ylabel("validation Loss")
plt.show()
```

```python
# Testing Accuracy and Loss
loss, accuracy = model.evaluate(x_test, y_test)
```

```python
# Test Predict
y_pred = model.predict(x_test)
```
